```python

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt



from pandas.plotting import scatter_matrix
from completejourney_py import get_data

cj_data = get_data()
cj_data.keys()

cj_data['transactions'].head()
transactions = cj_data['transactions']

products = cj_data['products']
ab =  pd.merge(transactions,products,on='product_id')
ab


ab.columns


# Average sales of each product type and the quantity

aa_1 = ab.groupby(['quantity','product_type'], as_index=True).agg({'sales_value': 'mean'}).nlargest(5, 'sales_value')
aa_1


#sns.jointplot(x='quantity', y='sales_value', data = aa_1, kind='reg', palette='YlGnBu')

aa_1.plot.barh(stacked=True , color= 'pink' , title = "Average sales of the product and quantity" )
```
