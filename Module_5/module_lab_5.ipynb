```python
import numpy as np
import pandas as pd 
import matplotlib .pyplot as plt
import mpld3

from pandas.plotting import scatter_matrix
from completejourney_py import get_data

cj_data = get_data()
cj_data.keys()

cj_data['transactions'].head()
transactions = cj_data['transactions']
demographics = cj_data['demographics']
products = cj_data['products']

promotions = cj_data['promotions']

coupon_redemp =  cj_data['coupon_redemptions']

coupons = cj_data['coupons']

campaign_descriptions = cj_data['campaign_descriptions']

campaigns = cj_data['campaigns']
```

```python
transactions

products
promotions

transactions.columns.intersection(campaigns.columns)
transactions.columns.intersection(demographics.columns)

# 1  -- Total sales for each household and the income range as per the household composition

new_df= pd.merge(transactions,demographics,on='household_id')
new_df.columns


# aa = new_df.groupby(['income','household_comp'], as_index=False).agg({'sales_value': 'sum'}).nlargest(10000000, 'sales_value')

aa = new_df.groupby(['income','household_comp'], as_index=False).agg({'sales_value': 'sum'}).nlargest(10, 'sales_value')
aa

#aa.plot.scatter(x='income',y='sales_value' , title= 'sales and household comp')
plt.subplots(figsize=(25,10 ))
plt.subplot(221)
plt.plot('sales_value','income',data=aa , marker='o' , alpha=0.4)
plt.title('Total sales value for households with the income range', fontsize=12, color='darkslategray', loc='left', style='italic')
plt.legend()
plt.xlabel('Sales Value')
plt.ylabel('Income')

plt.subplot(222)
plt.plot( 'sales_value','household_comp', data=aa, linestyle='none', marker='o', color="orange", alpha=0.3)
plt.title('Total sales value for households with household composition', fontsize=12, color='darkslategray', loc='left', style='italic')
plt.legend()
plt.xlabel('Sales Value')
plt.ylabel('Household Composition')


plt.subplot(223)
plt.plot( 'sales_value','household_comp', data=aa, linestyle='none', marker='D', color="seagreen", alpha=0.3)
plt.title('Total sales value for households with the household composition', fontsize=12, color='darkslategray', loc='left', style='italic')
plt.legend()
plt.xlabel('Sales Value')
plt.ylabel('Household Composition')

plt.subplot(224)
plt.plot( 'sales_value','income', data=aa, marker='o', color="purple", alpha=0.3)
plt.title('Total sales value for households with the income range', fontsize=12, color='darkslategray', loc='left', style='italic')
plt.legend()
plt.xlabel('Sales Value')
plt.ylabel('Income')

# Add a title:
plt.suptitle('Total sales value for households with the income range and the household composition', y=1.02)


plt.show()



```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})




transactions.columns.intersection(products.columns)
transactions.columns.intersection(demographics.columns)

transactions.columns.intersection(campaigns.columns)
transactions.columns.intersection(demographics.columns)

# 1  -- Total sales for each household and the income range as per the household composition

new_df_2= pd.merge(transactions,demographics,on='household_id')
new_df_2.columns




new_df_2_1 = pd.merge(new_df_2,products,on='product_id')
new_df_2_1.columns

new_df_2_1

# aa = new_df.groupby(['income','household_comp'], as_index=False).agg({'sales_value': 'sum'}).nlargest(10000000, 'sales_value')

second_gr = new_df_2_1.groupby(['product_category','retail_disc'], as_index=False).agg({'retail_disc': 'sum'}).nlargest(50, 'retail_disc')
second_gr

sns.scatterplot(x='product_category', y='retail_disc', data = second_gr)


# Initialize the FacetGrid object
g = np.tile(list("product_category"), 50)
pal = sns.cubehelix_palette(10, rot=-.25, light=.7)
g = sns.FacetGrid(second_gr, row="product_category", aspect=15, height=.5, palette=pal)


g.map(sns.kdeplot,"retail_disc", bw_adjust=.5, clip_on=False,fill=True, alpha=1, linewidth=1.5)

g.map(sns.kdeplot, "retail_disc", clip_on=False, color="w", lw=2, bw_adjust=.5)

g.refline(y=0, linewidth=2, linestyle="-", color=None, clip_on=False)


def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)
```
