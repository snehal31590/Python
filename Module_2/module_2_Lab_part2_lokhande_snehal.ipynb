```python
import pandas as pd

url ='https://academic.udayton.edu/kissock/http/Weather/gsod95-current/OHCINCIN.txt'
weather = pd.read_csv(url, delimiter="\s+", header=None)
weather
weather.columns = ['Month', 'Day', 'Year', 'Avg_temp']

print('The dimensions of this data is' ,weather.shape)

#Missing Values are not present
weather.isnull()

# 365th row data 
third_six_fiv_row = weather.loc[364]
print('Observation for the 365th row in weather data frame' ,third_six_fiv_row)


# Extract a column and first 31 days in the column
new_avg_temp = weather['Avg_temp']
new_avg_temp
new_avg_temp_1 = new_avg_temp.head(31)
new_avg_temp_1.describe()

print('The output of the Series or DataFrame is',type(new_avg_temp))

print('Summary statistics for the dataframe new_avg_temp are --',new_avg_temp_1.describe())
```
